## What is GraphRAG?

GraphRAG (Graph Retrieval-Augmented Generation) is a method that combines:

- **Knowledge Graphs**: Structured representation of information with entities and relationships
- **Retrieval**: Finding relevant information from the graph based on queries
- **Augmented Generation**: Using retrieved graph information to enhance LLM responses

### Key Advantages:
- **Better Context**: Structured relationships provide richer context than raw text
- **Entity Awareness**: Understands connections between entities
- **Explainability**: Relationships make reasoning transparent
- **Scalability**: Efficient querying on large knowledge graphs
- **Semantic Understanding**: Captures meaning through relationships

## Step 2: Query Your Knowledge Graph

Once data is in Neo4j, you can query it to retrieve relevant information. Common query patterns include:

````python
# Query 1: Find entities related to a person
query_result_1 = graph.query("""
MATCH (p:Person {id: "Elon Musk"})-[r]->(related)
RETURN p.id as person, type(r) as relationship, related.id as target
LIMIT 10
""")

print("Entities related to Elon Musk:")
for record in query_result_1:
    print(f"  {record['person']} --[{record['relationship']}]--> {record['target']}")
````

## Step 3: Build a GraphRAG Chain

Create a retrieval-augmented generation chain that uses the knowledge graph:

1. **User Query** → 2. **Retrieve from Graph** → 3. **LLM with Context** → 4. **Generated Answer**

````python
from langchain_classic.prompts import ChatPromptTemplate
from langchain_classic.chains import GraphCypherQAChain

# Create a prompt template for the RAG chain
prompt_template = """
Use the following context from a knowledge graph to answer the question.

Context:
{context}

Question: {question}

Answer:
"""

# Simple GraphRAG implementation
def graph_rag_query(user_query: str):
    # Step 1: Retrieve relevant information from the graph
    retrieval_query = f"""
    MATCH (n)-[r]->(m)
    WHERE toLower(n.id) CONTAINS toLower("{user_query}") 
       OR toLower(m.id) CONTAINS toLower("{user_query}")
    RETURN DISTINCT n.id + ' ' + type(r) + ' ' + m.id as context
    LIMIT 5
    """
    
    graph_context = graph.query(retrieval_query)
    context_text = "\n".join([record['context'] for record in graph_context]) if graph_context else "No specific information found"
    
    # Step 2: Use LLM with graph context
    messages = [
        {
            "role": "system",
            "content": f"Use this knowledge graph information to answer questions accurately:\n{context_text}"
        },
        {
            "role": "user",
            "content": user_query
        }
    ]
    
    response = llm.invoke(messages)
    return response.content

# Test the GraphRAG chain
test_query = "Tell me about Elon Musk's work"
print("Query:", test_query)
print("\nResponse:")
print(graph_rag_query(test_query))
````

## Step 4: Advanced Graph Query Patterns

Learn common Cypher patterns for effective retrieval:

````python
# Pattern 1: Multi-hop queries (find indirect relationships)
print("=== Pattern 1: Multi-hop Relationships ===")
multi_hop_query = """
MATCH (a:Person)-[r1]->(b)-[r2]->(c)
RETURN a.id as source, r1, b.id as intermediate, r2, c.id as target
LIMIT 5
"""
result = graph.query(multi_hop_query)
print(f"Found {len(result)} multi-hop relationships\n")

# Pattern 2: Entity-centric queries (all relationships for an entity)
print("=== Pattern 2: Entity Profile ===")
entity_query = """
MATCH (p:Person {id: "Elon Musk"})-[r]-(n)
WITH type(r) as relationship_type, n, collect(n) as nodes
RETURN relationship_type, count(n) as count
"""
result = graph.query(entity_query)
print("Relationship types for Elon Musk:")
for record in result:
    print(f"  {record['relationship_type']}: {record['count']} connections")

# Pattern 3: Dense subgraph (cluster of related entities)
print("\n=== Pattern 3: Find Related Entities Cluster ===")
cluster_query = """
MATCH (n:Organization)-[r1]-(p:Person)-[r2]-(c:Concept)
WHERE EXISTS((n)--(p)--(c))
RETURN DISTINCT n.id, p.id, c.id
"""
result = graph.query(cluster_query)
print(f"Found {len(result)} related entity triples")
print("Sample related entities:")
for record in result:
    print(f"  Organization: {record['n.id']}, Person: {record['p.id']}, Concept: {record['c.id']}")
````

## Step 5: Best Practices for GraphRAG Projects

### 1. **Schema Design**
- Define clear node labels (Person, Organization, Concept, Location, etc.)
- Create meaningful relationship types (WORKS_FOR, LOCATED_IN, CREATED, etc.)
- Add properties to both nodes and relationships with consistent naming

### 2. **Data Quality**
- Remove duplicate entities (entity linking/deduplication)
- Validate relationships for correctness
- Keep property values consistent (dates, formats)
- Use unique identifiers where possible

### 3. **Performance Optimization**
- Create indexes on frequently queried properties
- Limit query depth and result size
- Use APOC procedures for complex operations
- Monitor query execution time

### 4. **Query Strategy**
- Start with entity identification
- Use relationship patterns to expand context
- Combine graph queries with vector similarity for semantic search
- Cache frequently accessed subgraphs

### 5. **RAG Chain Design**
- Retrieve more context than you immediately need (allows flexible generation)
- Include relationship context, not just entity names
- Provide type information (is this a Person, Organization, etc.?)
- Use relevance scoring to rank retrieved information

### 6. **Iterative Improvement**
- Test with real queries and user feedback
- Refine your extraction patterns based on results
- Expand the knowledge graph incrementally
- Monitor LLM response quality

## Step 6: Implementing Schema & Indexes

Set up your Neo4j database for optimal performance:

````python
# Create indexes for faster queries
indexing_queries = [
    # Create unique constraints (also creates indexes)
    "CREATE CONSTRAINT unique_person_name IF NOT EXISTS FOR (p:Person) REQUIRE p.name IS UNIQUE",
    "CREATE CONSTRAINT unique_org_name IF NOT EXISTS FOR (o:Organization) REQUIRE o.name IS UNIQUE",
    
    # Create regular indexes for frequently searched properties
    "CREATE INDEX idx_person_name IF NOT EXISTS FOR (p:Person) ON (p.name)",
    "CREATE INDEX idx_org_name IF NOT EXISTS FOR (o:Organization) ON (o.name)",
    "CREATE INDEX idx_concept_name IF NOT EXISTS FOR (c:Concept) ON (c.name)",
]

print("Creating indexes and constraints...")
for query in indexing_queries:
    try:
        graph.query(query)
        print(f"✓ {query.split()[1:4]}")
    except Exception as e:
        print(f"Note: {str(e)[:60]}...")

print("\n✓ Database schema optimized!")
````

## Step 7: Building a Real GraphRAG Application

### Architecture for Production GraphRAG System:

```
Documents/PDFs → Extraction → Graph Transformation → Store in Neo4j
                                                           ↓
                                                     Knowledge Graph
                                                           ↓
User Query → Entity Extraction → Graph Query → Retrieve Context → LLM → Answer
```

### Key Components to Implement:

1. **Document Loader**: Load multiple documents (PDFs, web pages, APIs)
2. **Entity Extractor**: Identify and extract entities from text
3. **Relationship Extractor**: Find relationships between entities
4. **Deduplication**: Handle entity linking and consolidation
5. **Graph Query Engine**: Efficient retrieval from Neo4j
6. **LLM Prompt Engineering**: Optimize context-aware prompts
7. **Evaluation**: Measure answer quality and retrieval effectiveness

````python
# Example: Processing Multiple Documents
documents_batch = [
    Document(page_content="Steve Jobs founded Apple in 1976. He pioneered personal computing."),
    Document(page_content="Tesla manufactures electric vehicles. Elon Musk leads the company."),
    Document(page_content="Apple and Tesla compete in the tech innovation space."),
]

# Transform all documents
print("Processing batch of documents...")
all_graph_documents = []

for i, doc in enumerate(documents_batch):
    graph_docs = llm_transformer.convert_to_graph_documents([doc])
    all_graph_documents.extend(graph_docs)
    print(f"✓ Document {i+1}: {len(graph_docs[0].nodes)} entities, {len(graph_docs[0].relationships)} relationships")

# Store in database
print("\nStoring all documents in Neo4j...")
graph.add_graph_documents(all_graph_documents)
print("✓ All documents stored successfully!")

# Retrieve statistics
stats_query = """
MATCH (n)
RETURN count(*) as total_nodes
"""
total_nodes = graph.query(stats_query)[0]['total_nodes']
print(f"\nTotal entities in knowledge graph: {total_nodes}")
````

## Step 8: Common Challenges & Solutions

### Challenge 1: Entity Deduplication
**Problem**: Same entity extracted with different names/variations
```
"Elon Musk" vs "E. Musk" vs "Elon R. Musk"
```
**Solution**:
- Use entity linking algorithms
- Implement fuzzy string matching
- Maintain an entity alias mapping
- Use NER post-processing to merge duplicates

### Challenge 2: Relationship Extraction Quality
**Problem**: Incorrect or vague relationships
**Solution**:
- Use more specific LLM models for extraction
- Implement confidence scoring
- Validate relationships against domain knowledge
- Use multiple extraction passes

### Challenge 3: Graph Size & Complexity  
**Problem**: Graph becomes too large for efficient querying
**Solution**:
- Implement hierarchical summarization
- Use graph sampling techniques
- Create subgraph snapshots around key entities
- Implement time-based partitioning for temporal data

### Challenge 4: Context Relevance
**Problem**: Retrieved context doesn't match user intent
**Solution**:
- Combine graph retrieval with semantic search (embeddings)
- Rank results by relevance scores
- Implement multi-step retrieval strategies
- Use query expansion and reformulation
